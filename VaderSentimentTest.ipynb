```python
f = open('log','r')
text = f.read()
text = text.strip()

print("File is {} bytes large \n".format(len(text)))
# print(text)
sentences = text.split("Start_of_the_tweet: ")[1:]

print("File has {} sentences \n".format(len(sentences)))
# print(sentences)
```

```python
# sentences.extend(tricky_sentences)
sid = SentimentIntensityAnalyzer()
for sentence in sentences[:100]:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):
        print('{0}: {1}, '.format(k, ss[k]), end='')
    print("\n\n")
```
